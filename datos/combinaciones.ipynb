```python
import time
from itertools import combinations
from multiprocessing import Pool

# Función que calcula las tripletas cubiertas por una combinación de 6 números
def calcular_tripletas_cubiertas(combinacion_6, tripletas_no_cubiertas):
    tripletas_en_combinacion = set(combinations(combinacion_6, 3))
    tripletas_cubiertas = tripletas_en_combinacion & tripletas_no_cubiertas
    return combinacion_6, len(tripletas_cubiertas)

# Paso 1: Genera todas las combinaciones de 3 números de un conjunto de 30
numeros = list(range(1, 31))  # Del 1 al 30
tripletas = list(combinations(numeros, 3))  # Todas las combinaciones de 3 números
tripletas_no_cubiertas = set(tripletas)  # Usamos un set para facilitar el seguimiento

# Lista para almacenar las combinaciones de 6 números que vamos a construir
combinaciones_de_6 = []

# Inicia el tiempo
start_time = time.time()

# Paso 2: Itera mientras haya tripletas no cubiertas
while tripletas_no_cubiertas:
    mejor_combinacion = None
    max_tripletas_cubiertas = 0

    # Paso 3: Usa multiprocessing para calcular las tripletas cubiertas de manera paralela
    with Pool() as pool:
        resultados = pool.starmap(calcular_tripletas_cubiertas, 
                                 [(combinacion_6, tripletas_no_cubiertas) for combinacion_6 in combinations(numeros, 6)])
    
    # Paso 4: Encuentra la mejor combinación
    for combinacion_6, tripletas_cubiertas in resultados:
        if tripletas_cubiertas > max_tripletas_cubiertas:
            mejor_combinacion = combinacion_6
            max_tripletas_cubiertas = tripletas_cubiertas

    # Paso 5: Agrega la mejor combinación a la lista final y actualiza tripletas no cubiertas
    combinaciones_de_6.append(mejor_combinacion)
    tripletas_no_cubiertas -= set(combinations(mejor_combinacion, 3))

    # Imprime la combinación de 6 números que se acaba de agregar y el progreso
    print(f"Combinación añadida: {mejor_combinacion}")
    print(f"Tripletas restantes por cubrir: {len(tripletas_no_cubiertas)}")

# Resultado final
end_time = time.time()

print(f"\nCombinaciones necesarias para cubrir todas las tripletas: {len(combinaciones_de_6)}")
print("Combinaciones de 6 números:", combinaciones_de_6)
print(f"Tiempo de ejecución: {end_time - start_time:.2f} segundos")
```
